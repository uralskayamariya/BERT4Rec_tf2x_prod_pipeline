```python
from collections import defaultdict
```

```python
def data_partition(fname):
    usernum = 0
    itemnum = 0
    User = defaultdict(list)
    user_train = {}
    user_valid = {}
    user_test = {}
    # assume user/item index starting from 1
    with open(fname, 'r') as f:
        lines = list(f.readlines())
        for line in lines:
            line = line.rstrip()
            if ' ' not in line:
                continue
            line_split = line.rsplit(' ')
            u = int(line_split[0])
            i = int(line_split[1])
            usernum = max(u, usernum)
            itemnum = max(i, itemnum)
            User[u].append(i)

        for user in User:
            nfeedback = len(User[user])
            if nfeedback < 3:
                user_train[user] = User[user]
                user_valid[user] = []
                user_test[user] = []
            else:
                user_train[user] = User[user][:-2]
                user_valid[user] = []
                user_valid[user].append(User[user][-2])
                user_test[user] = []
                user_test[user].append(User[user][-1])
    return [user_train, user_valid, user_test, usernum, itemnum]
```

```python
[user_train, user_valid, user_test, usernum, itemnum] = data_partition('../data/interim/ml-1m.txt')
```

# Создание raw датафрейма

```python
import pandas as pd
```

```python
list_uid = []
list_sid = []
for line in lines:
    uid = int(line.split()[0])
    sid = int(line.split()[1].split('\n')[0])
    list_uid.append(uid)
    list_sid.append(sid)

df = pd.DataFrame({'uid': list_uid, 'sid_add_to_cart': list_sid})
df
```

```python
# группируем по uid и объединяем значения sid в список
grouped_df = df.groupby('uid')['sid_add_to_cart'].apply(list).reset_index()

print(grouped_df)
```

```python
import pickle
```

```python
def save_pkl(path, df):
    with open(path,'wb') as out:
        pickle.dump(df, out)
```
